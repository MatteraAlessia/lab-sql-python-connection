```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:YOUR_PASSWORD@localhost/sakila")
```

```python
def rentals_month(engine, month, year):
    query = f"""
    SELECT *
    FROM rental
    WHERE MONTH(rental_date) = {month}
      AND YEAR(rental_date) = {year}
    """
    return pd.read_sql(query, engine)
```

```python
def rental_count_month(df, month, year):
    col_name = f"rentals_{month}_{year}"
    
    result = df.groupby("customer_id").size().reset_index()
    result.columns = ["customer_id", col_name]
    
    return result
```

```python
def compare_rentals(df1, df2):
    merged = pd.merge(df1, df2, on="customer_id")
    
    merged["difference"] = merged.iloc[:, 2] - merged.iloc[:, 1]
    
    return merged
```
